```python
import numpy as np
import matplotlib.pylab as plt
from scipy.fftpack import fft, fftfreq
```

```python
N = 500
T = 1.3
f = 0.769230769
x = np.linspace(0.0, N*T, N)
y = np.sin(f * 2.0 * np.pi * x)

yf = fft(y)
xf = fftfreq(yf.shape[0],T)

yf = np.fft.fftshift(yf)
xf = np.fft.fftshift(xf)

f, (ax1, ax2) = plt.subplots(figsize=(14,14), nrows=2)
ax1.plot(x, y)
ax2.plot(xf, yf)
plt.xlim(0.0, 0.03)
```

```python
N = 500
T = 1.3
f = 0.769230769
x = np.linspace(0.0, N*T, N)
y = np.sin(f * 2.0 * np.pi * x)


zp_y = np.zeros(5000-N)
zp_y[:N] = y

yf = fft(zp_y)
xf = fftfreq(yf.shape[0])

yf = np.fft.fftshift(yf)
xf = np.fft.fftshift(xf)

f, (ax1, ax2) = plt.subplots(figsize=(14,14), nrows=2)
ax1.plot(x, y)
ax2.plot(xf, np.abs(yf))
ax2.scatter(xf, np.abs(yf))
plt.xlim(0.0, 0.03)
```
